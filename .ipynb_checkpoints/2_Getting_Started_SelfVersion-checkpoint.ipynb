```python
with tf.Session() as sess:
    X_input = tf.constant(x_bias)
    Y = tf.constant(np.transpose([y]).astype(np.float32))
    weights = tf.Variable(tf.random_normal([2,1],0,0.1))
    
    tf.global_variables_initializer().run()
    
    Y_hat = tf.matmul(X_input, weights)
    yerror = tf.subtract(Y_hat, Y)
    
    loss = (1/2)*tf.reduce_sum(tf.multiply(yerror,yerror))
    
    gradient = tf.reduce_sum(tf.transpose(tf.multiply(X_input,yerror)),1, keep_dims=True)
    '''
        # Shape(X_input, yerror) = 50x2
        Transposed result: 2x50
        summing across axis = 1 = sum all columns = 2x1
        Shape of weights = 2x1
    '''
    update_weights = tf.assign_sub(weights, learn_rate*gradient) 
    #tf.assign_sub(x,y) = x = x-y
    
    
    for _ in range(epochs):
        sess.run(update_weights)
        losses.append(loss.eval())
        
    beta = weights.eval()
    Y_hat = Y_hat.eval()

fig.set_size_inches (10,4)
fig, ax = plt.subplots()
ax.scatter(x,y,alpha=0.75)
line_x_range = (-4,6)
ax.plot(line_x_range,[beta[0] + a * beta[1] for a in line_x_range],"g", alpha = 0.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
```
